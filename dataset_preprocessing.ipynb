```python
import numpy as np
import json
from metadata.create_metadata import create_metadata_datasets
from metadata.utils import NpEncoder
```

```python
datasets_dir = './datasets/'
dataset_name = 'fruit_plants'

# path to directory with supervisely annotation
ANNOTATION_DIR = datasets_dir + dataset_name + '/' + dataset_name + '/ann/'
# path and name for file with dataset statistics
STAT_FILE = datasets_dir + 'all_metadata.json'

all_metadata = create_metadata_datasets(ANNOTATION_DIR)

with open(STAT_FILE, "w") as write_file:
    json.dump(all_metadata, write_file, cls=NpEncoder)
```

```python
dataset_names = [
    'Cassava Leaf Disease',
    'Corn or Maize Leaf Disease',
    'flower_classification',
    'fruit_plants',
    'Herbarium',
    'Plant Pathology',
    'satellite_images',
    'Tomato detection',
    'Wild Edible Plants'
]
```

```python
datasets_dir = './datasets/'

for dataset_name in dataset_names:

    # path to directory with supervisely annotation
    ANNOTATION_DIR = datasets_dir + dataset_name + '/' + dataset_name + '/ann/'
    # path and name for file with dataset statistics
    STAT_FILE = datasets_dir + 'all_metadata.json'

    all_metadata = create_metadata_datasets(ANNOTATION_DIR)

    with open(STAT_FILE, "w") as write_file:
        json.dump(all_metadata, write_file, cls=NpEncoder)
        
```
